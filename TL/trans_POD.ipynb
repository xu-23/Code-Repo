```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  # 更新为回归模型
from sklearn.metrics import mean_squared_error

# 1. 加载数据集A和数据集B
# 假设数据集A和B是DataFrame格式
data_A = pd.read_csv('ignition_delay_results.csv')  # 数据集A
data_B = pd.read_csv('ignition_delay_results_pure.csv')  # 数据集B

# print(data_A.iloc[10:16,:])  # 打印前5行数据，检查数据格式
# 假设特征在前，标签在最后一列
X_A = data_A.iloc[1:, :-1].values  # 特征
y_A = data_A.iloc[1:, -1].values    # 标签
X_B = data_B.iloc[1:, :-1].values     # 特征
y_B = data_B.iloc[1:, -1].values      # 标签

# print(X_B[10:16,:])  # 打印前5行数据，检查数据格式

# 2. 对数据集A进行PCA，提取主要特征
pca = PCA()  # 保留95%的方差
X_A_pca = pca.fit_transform(X_A)

# print(X_A_pca[10:16,:])  
# 3. 将数据集B转换为PCA特征空间
X_B_pca = pca.transform(X_B)

print(y_B[10:16])

```

```python
model0 = LinearRegression()  # 使用回归模型
# model0.fit(X_A_pca, y_A)

# 在数据集B上微调模型
model0.fit(X_B_pca, y_B)

# 5. 评估模型
y_B_pred = model0.predict(X_B_pca)
print(y_B_pred[10:16], y_B[10:16])
mse = mean_squared_error(y_B, y_B_pred)

print(f'在数据集B上的均方误差: {mse:.8f}')
```

```python
model = LinearRegression()  # 使用回归模型
model.fit(X_A_pca, y_A)

# 在数据集B上微调模型
# model.fit(X_B_pca, y_B)

# 5. 评估模型
y_B_pred = model.predict(X_B_pca)
print(y_B_pred[10:16], y_B[10:16])
mse = mean_squared_error(y_B, y_B_pred)

print(f'在数据集B上的均方误差: {mse:.8f}')
```
